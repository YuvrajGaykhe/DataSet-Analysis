```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("insurance (1).csv")


print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())

print("\nDescriptive Statistics (Numerical):")
print(df.describe())

print("\nDescriptive Statistics (Including Categorical):")
print(df.describe(include="all"))


print("\nMissing Values Per Column:")
print(df.isnull().sum())


print("\nNumber of Duplicate Rows:", df.duplicated().sum())


stats = df.describe().T
stats["variance"] = df.var(numeric_only=True)
stats["range"] = df.max(numeric_only=True) - df.min(numeric_only=True)

print("\nExtended Statistical Summary:")
print(stats)

plt.figure(figsize=(12,6))
df.select_dtypes(include="number").hist(bins=20)
plt.suptitle("Histogram Plot of Numerical Variables")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["age"], df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Insurance Charges")
plt.show()

plt.figure(figsize=(6,4))
df["smoker"].value_counts().plot(kind="bar")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.title("Smoker Distribution")
plt.show()


```
